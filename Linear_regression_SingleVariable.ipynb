```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv(r"C:\Users\Admin\Downloads\canada_per_capita_income (1).csv", sep='\t')
df
```

WE ARE PREDCTING THE PER CAPITA INCOME FOR THE YEAR


```python
%matplotlib inline
plt.xlabel('YEAR')
plt.ylabel('PER_CAPITA_INCOME')
reg= linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
plt.scatter(df.year, df.per_capita_income, color= 'red', marker='+')
```

```python
reg.predict([[2017]])
```
